under sampling
Build a sample dataset containing similar distribution of normal and fraudulent transaction

```python
legit1= legit.sample(n=492)
```

Concatenating two dataframe

```python
new_data = pd.concat([legit1,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
x = new_data.drop(columns='Class',axis =1)
y= new_data['Class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred_lreg=model.predict(x_test)
print(y_pred_lreg)
```

```python
from sklearn.metrics import classification_report , accuracy_score
print('classification MOdel')

print('--'*30)
logreg_accuracy = round (accuracy_score(y_test,y_pred_lreg)*100,2)
print('Accuracy', logreg_accuracy,'%')
```

```python
from sklearn.metrics import precision_score , recall_score, confusion_matrix

precision =precision_score(y_test ,y_pred_lreg)
recall = recall_score(y_test , y_pred_lreg)
print(f'precision: {precision:2f}')
print(f'recall :{recall:2f}')
confusion = confusion_matrix(y_test ,y_pred_lreg)
print(confusion)
sns.heatmap(confusion,annot =True ,fmt = "d")
```
